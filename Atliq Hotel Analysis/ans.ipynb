```python
df['month'] = df['check_in_date'].dt.strftime('%b %y')
df['week'] = df['check_in_date'].dt.strftime('%W')

def filter_data(df, properties=None, cities=None, statuses=None, platforms=None, months=None, weeks=None):
    filtered_df = df.copy()
    
    if properties:
        filtered_df = filtered_df[filtered_df['property_name'].isin(properties)]
    if cities:
        filtered_df = filtered_df[filtered_df['city'].isin(cities)]
    if statuses:
        filtered_df = filtered_df[filtered_df['booking_status'].isin(statuses)]
    if platforms:
        filtered_df = filtered_df[filtered_df['booking_platform'].isin(platforms)]
    if months:
        filtered_df = filtered_df[filtered_df['month'].isin(months)]
    if weeks:
        filtered_df = filtered_df[filtered_df['week'].isin(weeks)]
    
    return filtered_df
filtered_df = filter_data(df, 
                          properties=['Atliq Exotica', 'Atliq Grands'],
                          cities=['Mumbai'],
                          statuses=['Checked Out'],
                          platforms=['direct online'],
                          months=['May 22'],
                          weeks=['18', '19'])

print(filtered_df.shape)
filtered_df.head()
```

```python
data['check_in_date'] = pd.to_datetime(data['check_in_date'])

data['occupancy_rate'] = data['successful_bookings'] / data['capacity']

occupancy_by_hotel = data.groupby(['property_name', 'check_in_date'])['occupancy_rate'].mean().reset_index()

plt.figure(figsize=(20, 10))
sns.lineplot(data=occupancy_by_hotel, x='check_in_date', y='occupancy_rate', hue='property_name')
plt.title('Hotel Occupancy Rates Over Time')
plt.xlabel('Date')
plt.ylabel('Occupancy Rate')
plt.legend(title='Hotel', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### The occupancy rate is highest in the month of May and lowest in the month of June.

```python
data['check_in_date'] = pd.to_datetime(data['check_in_date'])

data['occupancy_rate'] = data['successful_bookings'] / data['capacity']

data['month_year'] = data['check_in_date'].dt.to_period('M')

occupancy_by_hotel_month = data.groupby(['property_name', 'month_year'])['occupancy_rate'].mean().reset_index()

occupancy_by_hotel_month['month_year'] = occupancy_by_hotel_month['month_year'].dt.to_timestamp()

plt.figure(figsize=(15, 8))
sns.lineplot(data=occupancy_by_hotel_month, x='month_year', y='occupancy_rate', hue='property_name', marker='o')
plt.title('Monthly Average Hotel Occupancy Rates')
plt.xlabel('Month')
plt.ylabel('Average Occupancy Rate')
plt.legend(title='Hotel', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Monthly Revenue by Each Hotel


```python
df['check_in_date'] = pd.to_datetime(df['check_in_date'])
df['month_year'] = df['check_in_date'].dt.to_period('M')
revenue_by_hotel_month = df.groupby(['property_name', 'month_year'])['revenue_realized'].sum().reset_index()
revenue_by_hotel_month['month_year'] = revenue_by_hotel_month['month_year'].dt.to_timestamp()
plt.figure(figsize=(15, 8))
sns.lineplot(data=revenue_by_hotel_month, x='month_year', y='revenue_realized', hue='property_name', marker='o')
plt.title('Monthly Revenue by Hotel')
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.legend(title='Hotel', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Monthly Ratings by Each Hotel


```python
df['check_in_date'] = pd.to_datetime(df['check_in_date'])
df['month_year'] = df['check_in_date'].dt.to_period('M')
rating_by_hotel_month = df.groupby(['property_name', 'month_year'])['ratings_given'].mean().reset_index()
rating_by_hotel_month['month_year'] = rating_by_hotel_month['month_year'].dt.to_timestamp()
plt.figure(figsize=(15, 8))
sns.lineplot(data=rating_by_hotel_month, x='month_year', y='ratings_given', hue='property_name', marker='o')
plt.title('Monthly Average Ratings by Hotel')
plt.xlabel('Month')
plt.ylabel('Average Rating')
plt.legend(title='Hotel', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Ratings vs Revenue by Each Hotel


```python
df['check_in_date'] = pd.to_datetime(df['check_in_date'])
df['month_year'] = df['check_in_date'].dt.to_period('M')

rating_revenue_by_hotel_month = df.groupby(['property_name', 'month_year']).agg({
    'ratings_given': 'mean',
    'revenue_realized': 'sum'
}).reset_index()

rating_revenue_by_hotel_month['month_year'] = rating_revenue_by_hotel_month['month_year'].dt.to_timestamp()

plt.figure(figsize=(15, 10))
sns.scatterplot(data=rating_revenue_by_hotel_month, 
                x='ratings_given', 
                y='revenue_realized', 
                hue='property_name', 
                size='month_year', 
                sizes=(20, 200),
                alpha=0.7)
plt.title('Ratings vs Revenue by Hotel')
plt.xlabel('Average Rating')
plt.ylabel('Total Revenue')
plt.legend(title='Hotel', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### Correlation between Ratings and Revenue by Each Hotel


```python
correlations = df.groupby('property_name').apply(lambda x: x['ratings_given'].corr(x['revenue_realized'])).sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x=correlations.index, y=correlations.values)
plt.title('Correlation between Ratings and Revenue by Hotel')
plt.xlabel('Hotel')
plt.ylabel('Correlation Coefficient')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

### Overall Trend: Ratings vs Revenue


```python
plt.figure(figsize=(10, 6))
sns.regplot(data=rating_revenue_by_hotel_month, x='ratings_given', y='revenue_realized')
plt.title('Overall Trend: Ratings vs Revenue')
plt.xlabel('Average Rating')
plt.ylabel('Total Revenue')
plt.tight_layout()
plt.show()
```

### Occupancy Rates by Booking Platform


```python
df['check_in_date'] = pd.to_datetime(df['check_in_date'])
df['occupancy'] = df.groupby(['property_name', 'check_in_date'])['booking_status'].transform(lambda x: (x == 'Checked Out').sum() / len(x))
occupancy_by_platform = df.groupby(['property_name', 'booking_platform', df['check_in_date'].dt.to_period('M')])['occupancy'].mean().reset_index()
occupancy_by_platform.rename(columns={'check_in_date': 'month'}, inplace=True)

plt.figure(figsize=(15, 8))
sns.boxplot(x='booking_platform', y='occupancy', data=occupancy_by_platform)
plt.title('Occupancy Rates by Booking Platform')
plt.xlabel('Booking Platform')
plt.ylabel('Occupancy Rate')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Total Revenue by Each City


```python
df['check_in_date'] = pd.to_datetime(df['check_in_date'])
df['month_year'] = df['check_in_date'].dt.to_period('M')
revenue_by_city_month = df.groupby(['city', 'month_year'])['revenue_realized'].sum().reset_index()
revenue_by_city_month['month_year'] = revenue_by_city_month['month_year'].dt.to_timestamp()
total_revenue_by_city = df.groupby('city')['revenue_realized'].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x=total_revenue_by_city.index, y=total_revenue_by_city.values)
plt.title('Total Revenue by City')
plt.xlabel('City')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
