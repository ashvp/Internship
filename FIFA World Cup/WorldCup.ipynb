```python
homeMatches = pd.read_sql_query("SELECT HomeTeamName, COUNT(HomeTeamName) as homeCount FROM worldcupmatches GROUP BY HomeTeamName ORDER BY homeCount DESC;",conn)
```

```python
homeMatches.head()
```

```python
awayMatches = pd.read_sql_query("SELECT AwayTeamName, COUNT(AwayTeamName) as awayCount FROM worldcupmatches GROUP BY AwayTeamName ORDER BY awayCount DESC;",conn)
```

```python
awayMatches.head()
```

```python
totalMatches = pd.merge(homeMatches,awayMatches,how='inner',left_on='hometeamname',right_on='awayteamname')
```

```python
totalMatches['totalCount'] = totalMatches['homecount'] + totalMatches['awaycount']
```

```python
totalMatches.head()
```

```python
sns.barplot(x=totalMatches['hometeamname'],y=totalMatches['totalCount'])
plt.xticks(rotation=90)
plt.show()
```

```python
homeWins = pd.read_sql_query("SELECT HomeTeamName, COUNT(HomeTeamName) as HomeWinCount FROM worldcupmatches WHERE HomeTeamGoals > AwayTeamGoals GROUP BY HomeTeamName ORDER BY HomeWinCount DESC;",conn)
```

```python
awayWins = pd.read_sql_query("SELECT AwayTeamName, COUNT(AwayTeamName) as AwayWinCount FROM worldcupmatches WHERE AwayTeamGoals > HomeTeamGoals GROUP BY AwayTeamName ORDER BY AwayWinCount DESC;",conn)
```

```python
homeWins.head()
```

```python
awayWins.head()
```

```python
totalWins = pd.merge(homeWins,awayWins,how='inner',left_on='hometeamname',right_on='awayteamname')
```

```python
totalWins['totalWinCount'] = totalWins['homewincount'] + totalWins['awaywincount']
```
